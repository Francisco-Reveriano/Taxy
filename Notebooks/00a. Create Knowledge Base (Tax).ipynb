```python
import os
import openai
from agents import Agent, Runner, ModelSettings, set_tracing_disabled, WebSearchTool, FileSearchTool
from openai.types.shared import Reasoning
from IPython.display import display, Markdown
import asyncio
set_tracing_disabled(True)
from dotenv import load_dotenv
from openai.types.responses import ResponseTextDeltaEvent
load_dotenv()
```

```python
# Initiate OpenAI Client
client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Initiate Vector Store in OpenAI Client
vs = client.vector_stores.create(name="Tax_Knowledge_Library")
print("Vector ID: {}".format(vs.id))

# Grab Files within the Folder
folder_path = '../Data/Knowledge_Base/Tax_Documentation'
file_path_dirs = []
for files in os.listdir(folder_path):
  if files.endswith(".pdf"):
    file_path_dirs.append(os.path.join(folder_path, files))
    print(files)

# Pass Files into Vector Database
file_streams = [open(p, "rb") for p in file_path_dirs]
try:
    batch = client.vector_stores.file_batches.upload_and_poll(
        vector_store_id=vs.id,
        files=file_streams,  # any number of local PDFs
    )
    print(batch.status, batch.file_counts)  # e.g., "completed" and counts
finally:
    for f in file_streams:
        f.close()
```
